```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load CSV file and plot them

file_path = 'filmlengthsratings.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Grouping the data by 'Year' to get the average 'Length_Minutes'
avg_film_length_by_year = df.groupby('Year')['Length_Minutes'].mean().reset_index()

# Creating the plot
plt.figure(figsize=(10,6))
plt.plot(avg_film_length_by_year['Year'], avg_film_length_by_year['Length_Minutes'], marker='o', linestyle='-', color='b')
plt.title('Average Film Length Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Film Length (Minutes)', fontsize=14)
plt.grid(True)
```

```python
def get_list(column_name):
    # Path to the CSV file
    csv_path = 'library_data.csv'
    
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(csv_path)
        
        # Check if the specified column exists in the DataFrame
        if column_name in df.columns:
            # Return the column data as a list
            return df[column_name].tolist()
        else:
            # If the column does not exist, raise an error
            raise ValueError("Column name not found in the DataFrame")
    except Exception as e:
        # Print any error that occurs during the process
        print(f"An error occurred: {e}")
```

```python
incomes = get_list("Total Income")
years = get_list("Year")
```

# Function

- Reusable block of code that we can use throughout the notebook.
- create_line_chart is the call function, (years, incomes) - argument value
- Using matplotlib.pyplot (plt) library module to generate linechart
- 

```python

def create_line_chart(years, incomes):
    plt.figure(figsize=(10, 6))  # Set the figure size to be larger for better readability
    plt.plot(years, incomes, marker='o', linestyle='-', color='dodgerblue')  # Plotting the data
    
    # Adding titles and labels
    plt.xlabel('Time')  # X-axis label as 'Time'
    plt.ylabel('')  # Y-axis label left blank as per your request
    
    # Style customizations
    plt.grid(True)  # Adding a grid for better readability of plotted data
    plt.tight_layout()  # Adjust layout to make room for label rotation
    
    # Display the plot
    plt.show()
```

```python
create_line_chart(years, expenses)
```

```python
csv_path = 'restaurant_data.csv'


def get_column_data(csv_path, column_name):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_path)

    # Check if the column exists in the DataFrame
    if column_name in df.columns:
        # Return the column data as a list
        return df[column_name].tolist()
    else:
        # If the column does not exist, raise an error
        raise ValueError("Column name not found in the DataFrame")

# Example usage:
scores = get_column_data(csv_path, "score")

print(len(scores))
```

```python
A = 90 #cutoff for restaurant rating
total = 0

for score in scores:
    if score >= A:
        total = total + 1

print(total)
```

```python
scores = [96, 91, 79, 93, 86]
A = 90
```

```python
a_scores = []  # store all of the scores that are an A

for score in scores:
    if score >= A:
        print("A")
        a_scores.append(score)

print(a_scores)
```

```python
scores = get_column_data(csv_path, "score")
A = 90
B = 80
C = 70
```

```python
grades = [] # append newly assigned grade with a loop

for score in scores:
    if score >= A:
        grades.append("A")
    elif score >= B:
        grades.append("B")
    elif score >= C:
        grades.append("C")
    else:
        grades.append("Failed")

print(len(grades))
```

```python
print("First ten scores:", scores[:10])
print("First ten grades:", grades[:10])
```
